```python
import numpy as np
import pandas as pd
from pathlib import Path
from skimage.io import imread
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```python
# set path & create labels
dir = Path("Data/EuroSAT")
df = pd.DataFrame({'path': list(dir.glob("**/*.jp*g"))})
labels=df['path'].astype(str).str.split('\\',expand=True)[3]
```

```python
# read in RGB images, save them in a data metrix
mat=np.zeros([27000,64,64,3])
for i,tmpPath in enumerate(df.path.values):
    img_rgb = imread(tmpPath)
    mat[i,:,:,:] = img_rgb
    
```

```python
# combine labels and random sampling
import random
random.seed(0)
# Here is a function to create a balanced dataset with n images in each NEW label
# If we keep Pasture as a label, n should not be greater than 2000.
def balanced_dataset_creator(n):
    selector=[]
    newlabels=[]
    
    selector=selector+random.sample(range(0, 3000), int(n/3))
    selector=selector+random.sample(range(16000, 18500), int(n/3))
    selector=selector+random.sample(range(14000, 16000), int(n/3))
    newlabels=newlabels+([0]*n) #'Agriculture'
    
    selector=selector+(random.sample(range(3000, 6000),n))
    newlabels=newlabels+([1]*n) #'Forest'
    
    selector=selector+(random.sample(range(6000, 9000),n))
    newlabels=newlabels+([2]*n) #'HerbaceousVegetation'
    
    selector=selector+(random.sample(range(9000, 11500),n))
    newlabels=newlabels+([3]*n) #'Highway'
    
    selector=selector+(random.sample(range(11500, 14000), int(n/2)))
    selector=selector+(random.sample(range(18500, 21500), int(n/2)))
    newlabels=newlabels+([4]*n)  #'Building'

    selector=selector+(random.sample(range(21500, 24000),  int(n/2)))
    selector=selector+(random.sample(range(24000, 27000),  int(n/2)))
    newlabels=newlabels+([5]*n) #'Water'
    
    newlabels = np.array(newlabels)
    
    return selector,newlabels
```

```python
selector,newlabels=balanced_dataset_creator(2400)
```

```python
#Scale the matrix
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
mat_b = mat_b.reshape([14400,12288])
X = scaler.fit_transform(mat_b).reshape([14400,64,64,3])
y = newlabels
```

```python
# split data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y)
```

```python
from __future__ import print_function
import keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras import backend as K
```

```python
batch_size = 500
num_classes = 6
epochs = 10
```

```python
y_train = y_train.astype("float32")
y_test = y_test.astype("float32")
y_train.dtype
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python

y_train.shape
```

```python
y_test.shape
```

```python
# convert class vectors to binary class matrices
y_train[1:20,:]
```

```python
X_train.shape
```

```python
# Model description
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=(64,64,3)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))

model.summary()
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model, show_shapes=True, show_layer_names=True)
```

```python
from tensorflow.keras.optimizers import Adadelta
model.compile(loss="categorical_crossentropy",
              optimizer=Adadelta(),
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, y_test))
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

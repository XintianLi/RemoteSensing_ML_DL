```python
eachlabel=pd.DataFrame(labels.value_counts()).reset_index().rename(columns={'index':'label',2:'count'}).sort_values(by='label')
eachlabel['range_end']=eachlabel['count'].cumsum()
eachlabel
```

```python
random.seed(0)
```

```python
# Here is a function to create a balanced dataset with n images in each NEW label
# If we keep Pasture as a label, n should not be greater than 2000.
def balanced_dataset_creator(n):
    selector=[]
    newlabels=[]
    
    selector=selector+random.sample(range(0, 3000), int(n/3))
    selector=selector+random.sample(range(16000, 18500), int(n/3))
    selector=selector+random.sample(range(14000, 16000), int(n/3))
    newlabels=newlabels+(['Agriculture']*n)
    
    selector=selector+(random.sample(range(3000, 6000),n))
    newlabels=newlabels+(['Forest']*n)
    
    selector=selector+(random.sample(range(6000, 9000),n))
    newlabels=newlabels+(['HerbaceousVegetation']*n)
    
    selector=selector+(random.sample(range(9000, 11500),n))
    newlabels=newlabels+(['Highway']*n)
    
    selector=selector+(random.sample(range(11500, 14000), int(n/2)))
    selector=selector+(random.sample(range(18500, 21500), int(n/2)))
    newlabels=newlabels+(['Building']*n)  

    selector=selector+(random.sample(range(21500, 24000),  int(n/2)))
    selector=selector+(random.sample(range(24000, 27000),  int(n/2)))
    newlabels=newlabels+(['Water']*n)
    
    return selector,newlabels
```

```python
selector,newlabels=balanced_dataset_creator(2400)
```

```python
mat_b=mat[selector]
```

```python
from sklearn.decomposition import PCA
```

```python
#PCA
pca = PCA(n_components=100)
```

```python
X_b_pca100=pca.fit_transform(X_b)
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title("PCA Cumulative Explained Variance Ratio By Principal Component")
```

```python
plt.plot(pca.explained_variance_ratio_[0:30])
plt.title("PCA Explained Variance Ratio By Principal Component")
```

```python
X_b_pca_30=X_b_pca100[:,0:30]
```

```python
X_b_pca_30.shape
```

```python
from PIL import Image
```

```python
plt.figure(figsize=(15,15))

for i in range(30):
    plt.subplot(6,5,i+1)
    p=pca.components_[i,:]
    p=p
    pImg=Image.fromarray((p*255).astype(np.uint8).reshape(64,64,3))
    plt.imshow(pImg)

plt.suptitle("First 30 PC, balanced dataset",fontsize=16,y=0.93)
plt.show()
```

```python
plt.figure(figsize=(15,15))

for i in range(30):
    plt.subplot(6,5,i+1)
    p=pca.components_[i,:]
    p=(p-p.min())/(p.max()-p.min())
    pImg=Image.fromarray((p*255).astype(np.uint8).reshape(64,64,3))
    plt.imshow(pImg)

plt.suptitle("First 30 PC standarized, balanced dataset",fontsize=16,y=0.93)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#Search for the best kernel and C value for svm
parameters = {"estimator__kernel":['sigmoid','rbf'], "estimator__C":[0.1,1, 10]}
svc = OneVsRestClassifier(SVC())
svc_grid = GridSearchCV(svc, parameters,cv=10)
svc_grid.fit(X_btr30, y_btr30)
```

```python
pd.DataFrame(svc_grid.cv_results_)[['params','mean_test_score','std_test_score','rank_test_score']].sort_values(by='rank_test_score')
```

```python
parameters2 = {"estimator__kernel":['rbf'], "estimator__C":[1, 10,15, 20, 30]}
svc2 = OneVsRestClassifier(SVC())
svc_grid2 = GridSearchCV(svc2, parameters2,cv=10)
svc_grid2.fit(X_btr30, y_btr30)
```

```python
pd.DataFrame(svc_grid2.cv_results_)[['params','mean_test_score','std_test_score','rank_test_score']].sort_values(by='rank_test_score')
```

```python
parameters3 = {"estimator__kernel":['rbf'], "estimator__C":[18, 23,25, 27]}
svc3 = OneVsRestClassifier(SVC())
svc_grid3 = GridSearchCV(svc3, parameters3,cv=10)
svc_grid3.fit(X_btr30, y_btr30)
```

```python
pd.DataFrame(svc_grid3.cv_results_)[['params','mean_test_score','std_test_score','rank_test_score']].sort_values(by='rank_test_score')
```

Note: It seems that C=20 is the best parameter here.
